```python
import random

def guess(x):
     random_number = random.randint(1,x)
     guess = 0

     while guess != random_number :
          guess = int(input(f"Guess a number between 1 and {x}"))

          if guess < random_number:
               print("Too low : guess little bit higher :")
          elif guess > random_number:
               print("Too high : guess little bit lower :")
     
     print(f"Congralution you have guessed the number {random_number} correctely.")

guess(10)
```

```python
# You think of number the computer will guess the number.


import random

def computer_guess(x):
     low = 1
     high = x
     feedback = ""
     while feedback != "c":
          if low != high:
               guess = random.randint(low, high)
          else:
               guess = low # could also be higher because low = high
          feedback = input(f"Is {guess} too high (H) too low (L), or correct (C) ??".lower())
          if feedback == "h":
               high = guess - 1
          elif feedback == "l":
               low = guess + 1
     print(f"Yeh! you have guessed the no {guess} correctely")        

computer_guess(1000)
     
```
